```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib as plt 
import matplotlib.pyplot as plt  
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
 df = load_iris()
```

```python
df
```

```python

pd.DataFrame(df['data'],columns=['sepal lemgth','sepal width','petal lenghth','petal width'])
```

```python
#indeopendent feature
X = pd.DataFrame(df['data'],columns=['sepal lemgth','sepal width','petal lenghth','petal width'])
```

```python
#dependent feature
y = df['target']
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size = 0.3,random_state=42)
```

```python
#apply decesion tree classifier
from sklearn.tree import DecisionTreeClassifier
treec=DecisionTreeClassifier()
```

```python
treec.fit(X_train,Y_train)
```

```python
##visualise the tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treec,filled=True)
```

```python
#it will make overfitting so post pruning and preprunning is done
#we will do post pruning
#apply decesion tree classifier
from sklearn.tree import DecisionTreeClassifier
treect=DecisionTreeClassifier(max_depth=2)
```

```python
treect.fit(X_train,Y_train)
```

```python
##visualise the post pruning tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treect,filled=True)
```

```python
y_pred = treect.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm = confusion_matrix(Y_test,y_pred)
print(cm)
print(classification_report(Y_test,y_pred))
```
